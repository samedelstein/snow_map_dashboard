## Tier 2: Collapse alert-level storms into **operational storm episodes**

Your `storms` / `storms_out` table is “meteorologically correct” (separated by alert *family*).  
For snow-ops analysis, you usually want *one* operational episode that merges overlapping (buffered) windows across families.

Rule:
- Sort by `ops_start`
- Start a new operational storm when `ops_start` is **after** the running max of prior `ops_end`


```python
import pandas as pd

storms_ops = storms.sort_values("storm_start").reset_index(drop=True).copy()

storms_ops["storm_end_running"] = storms_ops["storm_end"].cummax()
storms_ops["prev_storm_end_running"] = storms_ops["storm_end_running"].shift()

# Optional: allow small gaps to still be considered same episode
GAP_HOURS = 6
storms_ops["gap_hours"] = (
    (storms_ops["storm_start"] - storms_ops["prev_storm_end_running"])
    .dt.total_seconds()
    .div(3600)
)

storms_ops["new_op_storm"] = (
    storms_ops["prev_storm_end_running"].isna()
    | (storms_ops["gap_hours"] > GAP_HOURS)
)

storms_ops["op_storm_id"] = storms_ops["new_op_storm"].cumsum()
```

```python
severity_order = ["Minor", "Moderate", "Severe", "Extreme"]
storms_ops["_severity_cat"] = pd.Categorical(
    storms_ops["max_severity"], categories=severity_order, ordered=True
)

op_storms = (
    storms_ops.groupby("op_storm_id", as_index=False)
    .agg(
        op_storm_start=("storm_start", "min"),
        op_storm_end=("storm_end", "max"),
        event_families=("event_family", lambda x: ", ".join(sorted(set(x)))),
        alert_storms=("storm_key", "count"),
        peak_severity=("_severity_cat", "max"),
    )
)

# NOW apply the large ops buffers at the episode level
op_storms["ops_start"] = op_storms["op_storm_start"] - pd.Timedelta(hours=24)
op_storms["ops_end"]   = op_storms["op_storm_end"]   + pd.Timedelta(hours=48)

op_storms["op_storm_key"] = "op_storm_" + op_storms["op_storm_id"].astype(int).astype(str).str.zfill(2)
op_storms["op_storm_name"] = op_storms.apply(
    lambda r: f"Storm from {r['op_storm_start']:%Y-%m-%d %H:%M} to {r['op_storm_end']:%Y-%m-%d %H:%M} UTC",
    axis=1
)

op_storms = op_storms.sort_values("op_storm_start").reset_index(drop=True)
op_storms
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ----------------------------
# Config
# ----------------------------
STORM_START_UTC = pd.Timestamp("2025-12-28 14:01:00+00:00")
LOCAL_TZ = "US/Eastern"   # auto-handles EST/EDT correctly

# ----------------------------
# Load + parse
# ----------------------------
snapshot_df = pd.read_csv('https://raw.githubusercontent.com/samedelstein/snow_map_dashboard/refs/heads/main/data/snapshot_snow_routes/snapshots.csv')

snapshot_df["snapshot_ts"] = pd.to_datetime(
    snapshot_df["snapshot_ts"], utc=True, errors="coerce"
)

# Optional: de-dupe (prevents weird deltas if your pipeline ever repeats rows)
snapshot_df = snapshot_df.drop_duplicates(
    subset=["snowroutesegmentid", "snapshot_ts"],
    keep="last"
)

# Keep only needed columns
cols = ["snapshot_ts", "snowroutesegmentid", "passes", "segmentlength"]
all_df = snapshot_df[cols].dropna().copy()

# ----------------------------
# Baseline-safe delta computation (counts first storm jump correctly)
# ----------------------------
baseline = (
    all_df[all_df["snapshot_ts"] < STORM_START_UTC]
    .sort_values(["snowroutesegmentid", "snapshot_ts"])
    .groupby("snowroutesegmentid", as_index=False)
    .last()[["snowroutesegmentid", "passes"]]
    .rename(columns={"passes": "baseline_passes"})
)

df = (
    all_df[all_df["snapshot_ts"] >= STORM_START_UTC]
    .sort_values(["snowroutesegmentid", "snapshot_ts"])
    .merge(baseline, on="snowroutesegmentid", how="left")
)

df["baseline_passes"] = df["baseline_passes"].fillna(0)

# prev_passes: previous storm snapshot, or baseline for first storm row
df["prev_passes"] = df.groupby("snowroutesegmentid")["passes"].shift(1)
df["prev_passes"] = df["prev_passes"].fillna(df["baseline_passes"])

df["delta_passes"] = (df["passes"] - df["prev_passes"]).clip(lower=0)
df["delta_miles"] = df["delta_passes"] * df["segmentlength"]

# ----------------------------
# Convert timestamps to Eastern Time for analysis + display
# ----------------------------
df["snapshot_ts_et"] = df["snapshot_ts"].dt.tz_convert(LOCAL_TZ)

storm_start_et = STORM_START_UTC.tz_convert(LOCAL_TZ)

# Choose a stable "now" for repeatable results (set once per notebook)
if "EVAL_NOW_ET" not in globals():
    EVAL_NOW_ET = df["snapshot_ts_et"].max()
now_et = EVAL_NOW_ET

# ----------------------------
# Rolling window metrics in ET
# ----------------------------
def miles_in_last(hours: int) -> float:
    cutoff = now_et - pd.Timedelta(hours=hours)
    m = df.loc[
        (df["snapshot_ts_et"] >= cutoff) & (df["snapshot_ts_et"] <= now_et),
        "delta_miles"
    ].sum()
    return float(m)

metrics = {
    "past_1h": miles_in_last(1),
    "past_6h": miles_in_last(6),
    "past_12h": miles_in_last(12),
    "past_24h": miles_in_last(24),
    "since_storm_start": float(df.loc[df["snapshot_ts_et"] <= now_et, "delta_miles"].sum())
}

print("Using now (ET):", now_et)
print(metrics)

# ----------------------------
# Hourly miles plowed (last 24h), bucketed in ET (important!)
# ----------------------------
last24 = df.loc[
    (df["snapshot_ts_et"] >= now_et - pd.Timedelta(hours=24)) &
    (df["snapshot_ts_et"] <= now_et)
].copy()

last24["hour_et"] = last24["snapshot_ts_et"].dt.floor("h")

hourly = (
    last24.groupby("hour_et", as_index=False)["delta_miles"]
    .sum()
)

# Ensure continuous hourly axis
all_hours = pd.date_range(
    start=hourly["hour_et"].min(),
    end=hourly["hour_et"].max(),
    freq="h",
    tz=LOCAL_TZ
)

hourly = (
    hourly.set_index("hour_et")
          .reindex(all_hours, fill_value=0)
          .rename_axis("hour_et")
          .reset_index()
)

# ----------------------------
# NYT-style bar chart (Windows-safe formatter)
# ----------------------------
fig, ax = plt.subplots(figsize=(12.5, 4.3))

ax.bar(hourly["hour_et"], hourly["delta_miles"], width=0.035)

ax.set_title("Miles plowed per hour (last 24 hours)", loc="left", fontsize=14, pad=10)
ax.text(
    0, 1.02,
    f"Storm start: {storm_start_et:%Y-%m-%d %I:%M %p ET} • Latest snapshot: {now_et:%Y-%m-%d %I:%M %p ET}",
    transform=ax.transAxes, fontsize=10, va="bottom"
)

ax.set_ylabel("Miles plowed", fontsize=11)
ax.set_xlabel("")

ax.yaxis.grid(True, linewidth=0.6, alpha=0.25)
ax.set_axisbelow(True)

for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)

ax.tick_params(axis="y", length=0)
ax.tick_params(axis="x", length=0, labelsize=9)

ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%I %p\n%b %d"))  # Windows-safe

ax.yaxis.set_major_locator(plt.MaxNLocator(5))

# annotate peak
peak = hourly.loc[hourly["delta_miles"].idxmax()]
ax.annotate(
    f"Peak: {peak['delta_miles']:.0f}",
    xy=(peak["hour_et"], peak["delta_miles"]),
    xytext=(10, 8),
    textcoords="offset points",
    fontsize=9,
    arrowprops=dict(arrowstyle="-", linewidth=0.8, alpha=0.6)
)

ax.text(
    0, -0.18,
    "Miles computed from pass-count deltas × segment length; includes repeat passes (maintenance).",
    transform=ax.transAxes, fontsize=9, color="0.35"
)

# more robust than tight_layout on some setups
fig.set_constrained_layout(True)
plt.show()
```

```python
import pandas as pd

LOCAL_TZ = "US/Eastern"
storm_start_utc = pd.Timestamp("2025-12-28 14:01:00+00:00")

def segment_event_rows(snapshot_df: pd.DataFrame, seg_ids, storm_start=storm_start_utc):
    seg_ids = [seg_ids] if isinstance(seg_ids, str) else list(seg_ids)

    # Parse / keep needed cols
    cols = ["snapshot_ts", "snowroutesegmentid", "passes", "segmentlength"]
    all_df = snapshot_df[cols].copy()
    all_df["snapshot_ts"] = pd.to_datetime(all_df["snapshot_ts"], utc=True, errors="coerce")
    all_df = all_df.dropna(subset=["snapshot_ts", "snowroutesegmentid", "passes", "segmentlength"])
    all_df = all_df[all_df["snowroutesegmentid"].isin(seg_ids)].copy()

    # De-dupe exact segment+timestamp if your pipeline can repeat rows
    all_df = all_df.drop_duplicates(subset=["snowroutesegmentid", "snapshot_ts"], keep="last")

    # Baseline = last pass before storm start (per segment)
    baseline = (
        all_df[all_df["snapshot_ts"] < storm_start]
        .sort_values(["snowroutesegmentid", "snapshot_ts"])
        .groupby("snowroutesegmentid", as_index=False)
        .last()[["snowroutesegmentid", "passes"]]
        .rename(columns={"passes": "baseline_passes"})
    )

    # Storm rows
    df = (
        all_df[all_df["snapshot_ts"] >= storm_start]
        .sort_values(["snowroutesegmentid", "snapshot_ts"])
        .merge(baseline, on="snowroutesegmentid", how="left")
    )
    df["baseline_passes"] = df["baseline_passes"].fillna(0)

    # Previous pass: prior storm snapshot, otherwise baseline
    df["prev_passes"] = df.groupby("snowroutesegmentid")["passes"].shift(1)
    df["prev_passes"] = df["prev_passes"].fillna(df["baseline_passes"])

    # Deltas
    df["delta_passes"] = (df["passes"] - df["prev_passes"]).clip(lower=0)
    df["delta_miles"] = df["delta_passes"] * df["segmentlength"]

    # ET timestamps + ET hourly bucket (bucket AFTER tz_convert)
    df["snapshot_ts_et"] = df["snapshot_ts"].dt.tz_convert(LOCAL_TZ)
    df["hour_et"] = df["snapshot_ts_et"].dt.floor("h")

    return df
```

```python
def segment_hourly(snapshot_df, seg_ids, hours_back=24):
    events = segment_event_rows(snapshot_df, seg_ids)

    now_et = events["snapshot_ts_et"].max()
    cutoff = now_et - pd.Timedelta(hours=hours_back)

    hourly = (
        events[(events["snapshot_ts_et"] >= cutoff) & (events["snapshot_ts_et"] <= now_et)]
        .groupby(["snowroutesegmentid", "hour_et"], as_index=False)["delta_miles"]
        .sum()
        .sort_values(["snowroutesegmentid", "hour_et"])
    )

    return hourly, events.sort_values(["snowroutesegmentid", "snapshot_ts_et"])
```
