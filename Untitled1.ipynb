```python
df = pd.read_csv('https://raw.githubusercontent.com/samedelstein/snow_map_dashboard/refs/heads/main/data/snapshot_snow_routes/snapshots.csv')
```

```python
EVENT_START = pd.Timestamp("2025-12-26 18:18:15", tz="UTC")

segment_first_service = (
    df
    .dropna(subset=["lastserviced"])
    .loc[df["lastserviced"] > EVENT_START]
    .groupby("snowroutesegmentid")
    .agg(
        first_serviced=("lastserviced", "min"),
        snowrouteid=("snowrouteid", "first"),
        routepriority=("routepriority", "first"),
        roadname=("roadname", "first")
    )
    .reset_index()
)
```

```python
segment_first_service = segment_first_service.sort_values("first_serviced")

segment_first_service["rank"] = range(1, len(segment_first_service) + 1)
segment_first_service["total_segments"] = len(segment_first_service)
```

```python
NOW = df["snapshot_ts"].max()
NOW = pd.to_datetime(NOW, utc=True)  # harmless if already Timestamp

print("EVENT_START:", EVENT_START, type(EVENT_START))
print("NOW:", NOW, type(NOW))
segment_first_service["serviced_so_far"] = (
    segment_first_service["first_serviced"] <= NOW
)

Z = segment_first_service["serviced_so_far"].sum()
```

```python
def street_progress(segment_id):
    row = segment_first_service.loc[
        segment_first_service["snowroutesegmentid"] == segment_id
    ].iloc[0]

    return {
        "rank": int(row["rank"]),
        "total": int(row["total_segments"]),
        "serviced_so_far": int(Z),
        "remaining_before_you": int(row["rank"] - Z - 1)
    }
```

```python
segment_first_service["minutes_from_start"] = (
    segment_first_service["first_serviced"] - EVENT_START
).dt.total_seconds() / 60
```

```python
import pandas as pd
import numpy as np

df = df.copy()

# ensure datetime
df["lastserviced"] = pd.to_datetime(df["lastserviced"], utc=True, errors="coerce")

# define "serviced" (adjust if you have a better signal)
SERVICED = df["lastserviced"].notna()

```

```python

# keep a stable priority per segment (take most common non-null within event)
def mode_nonnull(s):
    s = s.dropna()
    return s.mode().iloc[0] if len(s) else np.nan

seg_event = (
    df.loc[SERVICED]
      .groupby(["event_id", "snowroutesegmentid"], as_index=False)
      .agg(
          first_serviced_at=("lastserviced", "min"),
          routepriority=("routepriority", mode_nonnull),
          snowrouteid=("snowrouteid", mode_nonnull),
          roadname=("roadname", mode_nonnull),
      )
)

# helpful: parse priority number (Priority 1/2/3 -> 1/2/3)
seg_event["priority_n"] = (
    seg_event["routepriority"]
      .astype(str)
      .str.extract(r"(\d+)")[0]
      .astype(float)
)
```

```python
seg_event
```

```python
seg_event["rank_in_event"] = (
    seg_event.groupby("event_id")["first_serviced_at"]
             .rank(method="dense", ascending=True)
             .astype(int)
)

# percentile: 0=earliest, 1=latest
seg_event["pct_in_event"] = (
    seg_event.groupby("event_id")["first_serviced_at"]
             .rank(method="average", pct=True, ascending=True)
)
```

```python
import numpy as np
import pandas as pd

def weighted_mean(x, w):
    x = np.asarray(x, dtype="float64")
    w = np.asarray(w, dtype="float64")
    m = np.isfinite(x) & np.isfinite(w)
    if m.sum() == 0:
        return np.nan
    return float((x[m] * w[m]).sum() / w[m].sum())

def mode_nonnull(s):
    s = s.dropna()
    if s.empty:
        return np.nan
    m = s.mode()
    return m.iloc[0] if not m.empty else s.iloc[0]
```

```python
priority_weight = {1: 3.0, 2: 2.0, 3: 1.0}
seg_event["w"] = seg_event["priority_n"].map(priority_weight).fillna(1.0)
```

```python
tmp = seg_event[["snowroutesegmentid", "pct_in_event", "w"]].copy()
tmp["xw"] = tmp["pct_in_event"] * tmp["w"]

global_pct = (
    tmp.groupby("snowroutesegmentid", as_index=False)
       .agg(sum_xw=("xw", "sum"), sum_w=("w", "sum"))
)
global_pct["typical_pct"] = global_pct["sum_xw"] / global_pct["sum_w"]
global_pct = global_pct.drop(columns=["sum_xw", "sum_w"])
```

```python
meta = (
    seg_event.groupby("snowroutesegmentid", as_index=False)
             .agg(
                 events_seen=("event_id", "nunique"),
                 median_priority=("priority_n", "median"),
                 roadname=("roadname", mode_nonnull),
                 snowrouteid=("snowrouteid", mode_nonnull),
                 routepriority=("routepriority", mode_nonnull),
             )
)

global_segment = (
    meta.merge(global_pct, on="snowroutesegmentid", how="left")
        .sort_values("typical_pct", ascending=True)
        .reset_index(drop=True)
)

global_segment["global_rank"] = np.arange(1, len(global_segment) + 1)
global_segment["global_pctile"] = global_segment["typical_pct"]  # 0 early â†’ 1 late
```
