```python
FEATURESERVER_URL = "https://services6.arcgis.com/bdPqSfflsdgFRVVM/ArcGIS/rest/services/Streets/FeatureServer"
LAYER_ID = 0  # change if needed
OUT_PATH = Path("streets.geojson")

# If you want to override server maxRecordCount, set e.g. 1000.
PAGE_SIZE = None  # None = use server maxRecordCount
SLEEP_SECS = 0.1
```

```python
def get_json(url, params=None, timeout=60):
    params = params or {}
    r = requests.get(url, params=params, timeout=timeout)
    r.raise_for_status()
    return r.json()

def get_layer_metadata(feature_server_url: str, layer_id: int):
    layer_url = f"{feature_server_url}/{layer_id}"
    meta = get_json(layer_url, {"f": "pjson"})
    max_record_count = int(meta.get("maxRecordCount", 2000))
    oid_field = meta.get("objectIdField") or meta.get("objectIdFieldName")
    if not oid_field:
        raise RuntimeError("Could not determine objectIdField from layer metadata.")
    return layer_url, oid_field, max_record_count
```

```python
def download_all_geojson(feature_server_url: str, layer_id: int, out_path: Path, page_size: int | None = None, sleep_secs: float = 0.0):
    layer_url, oid_field, max_record_count = get_layer_metadata(feature_server_url, layer_id)

    if page_size is None:
        page_size = max_record_count

    all_features = []
    offset = 0

    while True:
        params = {
            "where": "1=1",
            "outFields": "*",
            "returnGeometry": "true",
            "orderByFields": f"{oid_field} ASC",     # stable paging
            "resultOffset": offset,
            "resultRecordCount": page_size,
            "f": "geojson",
        }

        data = get_json(f"{layer_url}/query", params=params, timeout=120)

        features = data.get("features", [])
        all_features.extend(features)

        exceeded = data.get("exceededTransferLimit", False)

        print(f"offset={offset:<7} page={len(features):<5} total={len(all_features):<7} exceeded={exceeded}")

        # Stop conditions:
        # - If server says no more
        # - Or if page is shorter than page_size
        # - Or if it returned nothing
        if len(features) == 0:
            break
        if (not exceeded) and (len(features) < page_size):
            break

        offset += page_size
        if sleep_secs:
            time.sleep(sleep_secs)

    out_fc = {"type": "FeatureCollection", "features": all_features}
    out_path.write_text(json.dumps(out_fc))
    print(f"\n✅ Wrote {len(all_features)} features to: {out_path.resolve()}")

    return out_fc
```

```python
import re
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import LineString

streets_path = "streets.geojson"
winter_path  = "winter_operations_snow_routes_layer0.geojson"

streets = gpd.read_file(streets_path)
winter  = gpd.read_file(winter_path)

streets.crs, winter.crs, streets.shape, winter.shape
```

```python
def norm_name(s: str) -> str:
    if s is None:
        return ""
    s = str(s).upper().strip()
    s = re.sub(r"\s+", " ", s)
    # optional: normalize common suffix punctuation/spaces
    s = s.replace(".", "")
    return s

streets["name_norm"] = streets["FULLNAME"].map(norm_name)
winter["name_norm"]  = winter["roadname"].map(norm_name)

streets["name_norm"].nunique(), winter["name_norm"].nunique()
```

```python
streets_m = streets.to_crs(3857)
winter_m  = winter.to_crs(3857)
```

```python
import pandas as pd
import geopandas as gpd

BUFFER_METERS =  1.1 # try 30-40 if needed

street_keep = streets_m[[
    "FID", "FULLNAME", "name_norm",
    "BLK", "BLKSTREET",
    "L_F_ADD", "L_T_ADD", "R_F_ADD", "R_T_ADD",
    "BLOCK90_L", "BLOCK90_R",
    "geometry"
]].copy()

winter_keep = winter_m[[
    "OBJECTID", "snowroutesegmentid", "snowrouteid",
    "roadname", "name_norm",
    "geometry"
]].copy()

# ---- Make an explicit street index column that will survive sjoin
street_keep = street_keep.reset_index(drop=False).rename(columns={"index": "street_idx"})
# Build geometry lookup by street_idx
street_geom_lookup = street_keep.set_index("street_idx")["geometry"]

# ---- Buffer winter for candidate search
winter_buf = winter_keep.copy()
winter_buf["geometry"] = winter_buf.geometry.buffer(BUFFER_METERS)

# ---- Spatial join (candidate streets that intersect buffer)
cand = gpd.sjoin(
    winter_buf,
    street_keep,
    how="left",
    predicate="intersects"
).copy()

# Depending on geopandas, columns might be suffixed; handle both cases
left_name_col  = "name_norm_left"  if "name_norm_left"  in cand.columns else "name_norm"
right_name_col = "name_norm_right" if "name_norm_right" in cand.columns else "name_norm"

# If both sides ended up as the same col name (rare), keep both by renaming before join:
if left_name_col == right_name_col == "name_norm":
    # fallback: do join again with renamed columns to disambiguate
    winter_buf2 = winter_buf.rename(columns={"name_norm": "name_norm_w"})
    street_keep2 = street_keep.rename(columns={"name_norm": "name_norm_s"})
    cand = gpd.sjoin(winter_buf2, street_keep2, how="left", predicate="intersects").copy()
    left_name_col, right_name_col = "name_norm_w", "name_norm_s"

# ---- Filter to same normalized street name
cand = cand[cand[left_name_col] == cand[right_name_col]].copy()

# ---- Overlap scoring using ORIGINAL winter geometry (not buffered)
winter_geom = winter_keep.set_index("OBJECTID")["geometry"]

def overlap_len(row):
    # street_idx will exist even if index_right doesn't
    if pd.isna(row.get("street_idx")):
        return 0.0
    wgeom = winter_geom.loc[row["OBJECTID"]]
    sgeom = street_geom_lookup.loc[int(row["street_idx"])]
    inter = wgeom.intersection(sgeom)
    return inter.length if not inter.is_empty else 0.0

cand["overlap_m"] = cand.apply(overlap_len, axis=1)

# ---- Best match per winter OBJECTID
best = (
    cand.sort_values(["OBJECTID", "overlap_m"], ascending=[True, False])
        .drop_duplicates("OBJECTID")
        .copy()
)

# bring back the street attributes from street_keep using street_idx
best = best.merge(
    street_keep.drop(columns="geometry"),
    on="street_idx",
    how="left",
    suffixes=("", "_street")
)

best[["OBJECTID", "FID", "BLK", "BLKSTREET", "L_F_ADD", "L_T_ADD", "overlap_m"]].head()
```

```python
winter_enriched = winter_keep.merge(
    best[[
        "OBJECTID",
        "FID",
        "BLK", "BLKSTREET",
        "L_F_ADD", "L_T_ADD", "R_F_ADD", "R_T_ADD",
        "BLOCK90_L", "BLOCK90_R",
        "overlap_m"
    ]],
    on="OBJECTID",
    how="left"
)

winter_enriched_out = winter_enriched.to_crs(4326)
winter_enriched_out.to_file("winter_ops_with_blocks.geojson", driver="GeoJSON")
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd

# Candidate search radius (big = higher recall)
SEARCH_BUFFER_M = 50

# Scoring tolerance (small = how close is "close enough")
TOL_M = 10

street_keep = streets_m[[
    "FID", "FULLNAME", "name_norm",
    "BLK", "BLKSTREET",
    "L_F_ADD", "L_T_ADD", "R_F_ADD", "R_T_ADD",
    "BLOCK90_L", "BLOCK90_R",
    "geometry"
]].copy().reset_index(drop=False).rename(columns={"index": "street_idx"})

winter_keep = winter_m[[
    "OBJECTID", "snowroutesegmentid", "snowrouteid",
    "roadname", "name_norm",
    "geometry"
]].copy()

street_geom_lookup = street_keep.set_index("street_idx")["geometry"]
winter_geom_lookup = winter_keep.set_index("OBJECTID")["geometry"]

# --- Candidate generation
winter_buf = winter_keep.copy()
winter_buf["geometry"] = winter_buf.geometry.buffer(SEARCH_BUFFER_M)

cand = gpd.sjoin(
    winter_buf.rename(columns={"name_norm": "name_norm_w"}),
    street_keep.rename(columns={"name_norm": "name_norm_s"}),
    how="left",
    predicate="intersects",
).copy()

# Keep only same street name
cand = cand[cand["name_norm_w"] == cand["name_norm_s"]].copy()

# --- Scoring
def score_row(row):
    if pd.isna(row["street_idx"]):
        return pd.Series({"overlap_ratio": 0.0, "mean_dist": np.inf})

    w = winter_geom_lookup.loc[row["OBJECTID"]]
    s = street_geom_lookup.loc[int(row["street_idx"])]

    # fraction of winter segment that falls within TOL meters of street segment
    close_part = w.intersection(s.buffer(TOL_M))
    overlap_ratio = (close_part.length / w.length) if (not close_part.is_empty and w.length > 0) else 0.0

    # average-ish distance proxy (sample points along winter line)
    # (fast + stable without heavy computations)
    pts = [w.interpolate(t, normalized=True) for t in (0.25, 0.5, 0.75)]
    dists = [p.distance(s) for p in pts]
    mean_dist = float(np.mean(dists))

    return pd.Series({"overlap_ratio": overlap_ratio, "mean_dist": mean_dist})

cand[["overlap_ratio", "mean_dist"]] = cand.apply(score_row, axis=1)

# --- Pick best per winter OBJECTID:
# 1) maximize overlap_ratio
# 2) minimize mean_dist
best = (
    cand.sort_values(["OBJECTID", "overlap_ratio", "mean_dist"], ascending=[True, False, True])
        .drop_duplicates("OBJECTID")
        .copy()
)

# Optional: require minimum quality (prevents bad matches)
MIN_OVERLAP = 0.2  # tune: 0.1–0.5
best.loc[best["overlap_ratio"] < MIN_OVERLAP, ["street_idx"]] = np.nan

# Attach street attrs
best = best.merge(
    street_keep.drop(columns="geometry"),
    on="street_idx",
    how="left",
    suffixes=("", "_street")
)

best[["OBJECTID","snowroutesegmentid","roadname","FID","BLK","BLKSTREET","overlap_ratio","mean_dist"]].head(20)
```

```python
winter_enriched = winter_keep.merge(
    best[[
        "OBJECTID",
        "FID",
        "BLK", "BLKSTREET",
        "L_F_ADD", "L_T_ADD", "R_F_ADD", "R_T_ADD",
        "BLOCK90_L", "BLOCK90_R",
        "overlap_ratio", "mean_dist"
    ]],
    on="OBJECTID",
    how="left"
)
# --- After merge:
# winter_enriched = winter_keep.merge(... best ...)

import re
import numpy as np

# Roads where blocks/addresses generally don't make sense in the streets centerline file
NON_ADDRESSABLE_PATTERN = re.compile(
    r"\b(I\s?\d+|INTERSTATE|US\s?\d+|STATE\s?\d+|RAMP|EXIT|ENTRANCE|MEDIAN|SERVICE( RD)?|HWY|HIGHWAY|EXPY|EXPRESSWAY)\b",
    flags=re.IGNORECASE
)

winter_enriched["is_non_addressable"] = winter_enriched["roadname"].fillna("").apply(
    lambda s: bool(NON_ADDRESSABLE_PATTERN.search(s))
)

# For these, we *don't* want to treat missing BLK as a problem.
# Option A (recommended): keep the matched BLK columns but exclude them from QA
# Option B: blank them out so downstream never sees a block for these roads
BLANK_BLOCKS_FOR_NON_ADDRESSABLE = True

if BLANK_BLOCKS_FOR_NON_ADDRESSABLE:
    winter_enriched.loc[winter_enriched["is_non_addressable"], ["BLK", "BLKSTREET"]] = np.nan
    # optionally also clear address ranges
    winter_enriched.loc[winter_enriched["is_non_addressable"], ["L_F_ADD","L_T_ADD","R_F_ADD","R_T_ADD"]] = np.nan
PERFECT_OVERLAP = 0.90
MAX_GOOD_DIST = 10  # meters

if "overlap_ratio" in winter_enriched.columns and "mean_dist" in winter_enriched.columns:
    winter_enriched["qa_issue"] = (
        (winter_enriched["overlap_ratio"] < PERFECT_OVERLAP) |
        (winter_enriched["mean_dist"] > MAX_GOOD_DIST)
    )
else:
    # fallback if you only have overlap_m
    winter_enriched["qa_issue"] = winter_enriched["overlap_m"].fillna(0) <= 0

# Don't flag non-addressable roads as QA issues
winter_enriched.loc[winter_enriched["is_non_addressable"], "qa_issue"] = False
```

```python
import matplotlib.pyplot as plt
import contextily as ctx

def plot_match_debug(
    winter_enriched,
    winter_m,
    streets_m,
    *,
    blk=None,
    road_prefix=None,
    only_issues=True,
    take=10,
    pad=250,
    title=None
):
    sel = winter_enriched.copy()

    if only_issues and "qa_issue" in sel.columns:
        sel = sel[sel["qa_issue"]].copy()

    if blk is not None:
        sel = sel[sel["BLK"].astype(str) == str(blk)].copy()

    if road_prefix is not None:
        sel = sel[sel["roadname"].str.startswith(road_prefix, na=False)].copy()

    sel = sel.head(take).copy()

    cols = ["OBJECTID","snowroutesegmentid","roadname","BLK","FID"]
    if "overlap_ratio" in sel.columns: cols += ["overlap_ratio","mean_dist"]
    elif "overlap_m" in sel.columns: cols += ["overlap_m"]

    display(sel[cols])

    if sel.empty:
        print("No rows matched your filters.")
        return

    # --- 1) Combined/enriched geometry
    sel_combined_3857 = sel.to_crs(3857)

    # --- 2) Winter ops original geometry (same OBJECTID)
    obj_ids = sel["OBJECTID"].dropna().unique().tolist()
    sel_winter_orig_3857 = winter_m[winter_m["OBJECTID"].isin(obj_ids)].copy()

    # --- 3) Streets geometry (matched by FID)
    fids = sel["FID"].dropna().unique().tolist()
    sel_streets_3857 = streets_m[streets_m["FID"].isin(fids)].copy()

    fig, ax = plt.subplots(1, 1, figsize=(11, 11))

    # Streets
    if len(sel_streets_3857) > 0:
        sel_streets_3857.plot(ax=ax, color="green", linewidth=16, alpha=.45, label="Streets (source)")

    # Winter ops
    if len(sel_winter_orig_3857) > 0:
        sel_winter_orig_3857.plot(ax=ax, color="blue", linewidth=8, alpha=0.6, label="Winter ops (source)")

    # Combined
    if len(sel_combined_3857) > 0:
        sel_combined_3857.plot(ax=ax, color="red", linewidth=4, alpha=0.85, label="Combined (enriched)")

    # Zoom
    focus = sel_combined_3857 if len(sel_combined_3857) else sel_winter_orig_3857
    xmin, ymin, xmax, ymax = focus.total_bounds
    ax.set_xlim(xmin - pad, xmax + pad)
    ax.set_ylim(ymin - pad, ymax + pad)

    # Basemap
    ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

    ax.set_axis_off()
    ax.legend(loc="lower left")
    ax.set_title(title or "Segment comparison: Streets vs Winter Ops vs Combined")
    plt.show()
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

def inspect_objectid(
    oid: int,
    winter_enriched,
    winter_m,
    streets_m,
    cand=None,                 # optional: pass your candidate table if you still have it
    search_buffer_m=80,        # only used if cand is None
    tol_m=15,                  # only used if cand is None
    top_n=10,
    pad_m=250,
):
    # --- basic row
    row = winter_enriched[winter_enriched["OBJECTID"] == oid].copy()
    if row.empty:
        print(f"OBJECTID={oid} not found in winter_enriched")
        return

    # Print key fields
    cols = ["OBJECTID","snowroutesegmentid","snowrouteid","roadname","BLK","BLKSTREET","FID"]
    for c in ["overlap_ratio","mean_dist","overlap_m","qa_issue","is_non_addressable"]:
        if c in row.columns: cols.append(c)
    display(row[cols])

    # --- winter geometry
    w = winter_m[winter_m["OBJECTID"] == oid].copy()
    if w.empty:
        print("No geometry found in winter_m for this OBJECTID.")
        return
    wgeom = w.geometry.iloc[0]

    # --- chosen street geometry (by FID)
    chosen_fid = row["FID"].iloc[0]
    chosen = streets_m[streets_m["FID"] == chosen_fid].copy() if pd.notna(chosen_fid) else streets_m.iloc[0:0].copy()

    # --- get candidate streets
    if cand is not None and "OBJECTID" in cand.columns:
        cands = cand[cand["OBJECTID"] == oid].copy()
        if cands.empty:
            print("No cand rows for this OBJECTID; falling back to on-the-fly candidate search.")
            cand = None

    if cand is None:
        # On-the-fly candidates: spatial join within buffer, then score by overlap_ratio + mean_dist
        wbuf = gpd.GeoDataFrame(w.drop(columns="geometry"), geometry=w.geometry.buffer(search_buffer_m), crs=w.crs)
        cands = gpd.sjoin(wbuf, streets_m, how="left", predicate="intersects").copy()

        def score_street(sgeom):
            if sgeom is None or sgeom.is_empty:
                return 0.0, np.inf
            close_part = wgeom.intersection(sgeom.buffer(tol_m))
            overlap_ratio = (close_part.length / wgeom.length) if (not close_part.is_empty and wgeom.length > 0) else 0.0
            pts = [wgeom.interpolate(t, normalized=True) for t in (0.25, 0.5, 0.75)]
            mean_dist = float(np.mean([p.distance(sgeom) for p in pts]))
            return overlap_ratio, mean_dist

        scores = cands.geometry.apply(score_street)
        cands["overlap_ratio"] = [s[0] for s in scores]
        cands["mean_dist"] = [s[1] for s in scores]

    # show top candidates table
    show_cols = ["FID","FULLNAME","BLK","BLKSTREET"]
    for c in ["overlap_ratio","mean_dist","overlap_m"]:
        if c in cands.columns: show_cols.append(c)
    display(
        cands[show_cols]
        .dropna(subset=["FID"])
        .sort_values(["overlap_ratio","mean_dist"], ascending=[False, True] if "overlap_ratio" in cands.columns else True)
        .head(top_n)
    )

    # pick top N candidate geometries for plotting
    cands_top = (
        cands.dropna(subset=["FID"])
             .sort_values(["overlap_ratio","mean_dist"], ascending=[False, True] if "overlap_ratio" in cands.columns else True)
             .head(top_n)
             .copy()
    )

    # --- plot
    fig, ax = plt.subplots(1, 1, figsize=(11, 11))

    # candidates (gray)
    if len(cands_top) > 0:
        gpd.GeoSeries(cands_top.geometry, crs=streets_m.crs).plot(
            ax=ax, color="gray", linewidth=8, alpha=0.25, label=f"Top {top_n} candidates"
        )

    # chosen street (green)
    if len(chosen) > 0:
        chosen.plot(ax=ax, color="green", linewidth=10, alpha=0.65, label=f"Chosen street FID={chosen_fid}")

    # winter segment (red)
    w.plot(ax=ax, color="red", linewidth=4, alpha=0.95, label=f"Winter OBJECTID={oid}")

    # zoom + basemap
    xmin, ymin, xmax, ymax = w.total_bounds
    ax.set_xlim(xmin - pad_m, xmax + pad_m)
    ax.set_ylim(ymin - pad_m, ymax + pad_m)
    ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

    ax.set_axis_off()
    ax.legend(loc="lower left")
    ax.set_title(f"Inspect OBJECTID={oid}: {row['roadname'].iloc[0]}")
    plt.show()
```
